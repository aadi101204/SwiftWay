```python
import xml.etree.ElementTree as ET
tree = ET.parse('output_file.xml')
root = tree.getroot()
```

```python
#import numpy as np
#loss = 0
b=[]
for child in root:
    a = float(child.attrib['timeLoss'])
    #list = [a]
    b.append(a)
    #print(list,' ')
print(b)
```

```python
#Caculate average losstime
loss = 0
for child in root:
    a = float(child.attrib['timeLoss'])
    loss += a
ave_loss = loss / len(root)
print('ave_loss =', ave_loss)
print('%.2f'%ave_loss)
#Caculate average variance
difference = 0
for child in root:
    a = float(child.attrib['timeLoss'])
    difference += (a - ave_loss)**2
sigma = difference / len(root)
print('sigma =', sigma)
print('%.2f'%sigma)
print('The number of the vehicles is', len(root))

#Caculate average speed
sum_v = 0
for child in root:
    v = float(child.attrib['arrivalSpeed'])
    sum_v += v
ave_v = sum_v / len(root)
print('ave_v is', ave_v)
print('%.2f'%ave_v)
```
